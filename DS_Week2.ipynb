```python
#Mean/Mode Imputation
import pandas as pd
import numpy as np

# Sample dataset
df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

print("Original Data:")
print(df)

# Mean for numeric column
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Mode for categorical column
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])

print("\nAfter Handling Missing Values:")
print(df)
```

```python
#Forward Fill
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df_ffill = df.copy()

df_ffill.fillna(method='ffill', inplace=True)

print(df_ffill)
```

```python
#Backward Fill
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df_bfill = df.copy()

df_bfill.fillna(method='bfill', inplace=True)

print(df_bfill)
```

```python
#Deletion Method
   #Drop Rows
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df_drop_rows = df.dropna()

print("After dropping rows:")
print(df_drop_rows)
```

```python
#Deletion Method
   #Drop Columns
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df_drop_cols = df.dropna(axis=1)

print("After dropping columns:")
print(df_drop_cols)
```

```python
#REMOVING DUPLICATES

#Exact Match Removal
import pandas as pd

df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 4, 4],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'David', 'David'],
    'Age': [25, 30, 30, 35, 40, 40]
})

print("Original Data:")
print(df)

df_exact = df.drop_duplicates()

print("\nAfter Removing Exact Duplicates:")
print(df_exact)
```

```python
#Subset-Based Removal
import pandas as pd

df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 4, 4],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'David', 'David'],
    'Age': [25, 30, 30, 35, 40, 40]
})

df_subset_id = df.drop_duplicates(subset=['ID'])

print("\nAfter Removing Duplicate IDs:")
print(df_subset_id)

df_subset_name = df.drop_duplicates(subset=['Name'])

print("\nAfter Removing Duplicate Names:")
print(df_subset_name)
```

```python
#CORRECTING INCONSISTENT FORMATS
#1.Date Standardization
import pandas as pd

df = pd.DataFrame({
    'Date': ['2025-01-05', '05/01/2025', 'Jan 5, 2025', '2025.01.05']
})

df['Date'] = pd.to_datetime(
    df['Date'],
    errors='coerce'
).dt.strftime('%Y-%m-%d')

print(df)
```

```python
#2. Case Normalization
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'BOB', 'charlie', 'DAVID']
})

df['Name_lower'] = df['Name'].str.lower()

df['Name_upper'] = df['Name'].str.upper()

print(df)
```
